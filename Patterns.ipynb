```python
#hollow rectangle
for i in range(3):
  for j in range(5):
    if (i==0 or i==2 or j==0 or j==4):
      print("*" ,end=" ")
    else:
      print(" ", end=" ")
  print()
```

```python
#Half Pyramid

for i in range(6):
  for j in range(i+1):
    print("*", end= " ")
  print()
```

### Butterfly Pattern

```python
# Upper part of the butterfly
n = 5

for i in range(1, n + 1):
    # Left side of increasing stars
    for j in range(i):
        print("*", end=" ")

    # Spaces in the middle
    for j in range(2 * (n - i)):
        print(" ", end=" ")

    # Right side of increasing stars
    for j in range(i):
        print("*", end=" ")
    print()

for i in range(n, 0, -1):
    # Left side of decreasing stars
    for j in range(i):
        print("*", end=" ")

    # Spaces in the middle
    for j in range(2 * (n - i)):
        print(" ", end=" ")

    # Right side of decreasing stars
    for j in range(i):
        print("*", end=" ")
    print()
```
