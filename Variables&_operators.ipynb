```python
num1= int(input("Enter the first number:"))
num2= int(input("Enter second number:"))
print("Press 1 for addition")
print("Press 2 for subtraction")
print("Press 3 for multiplication")
print("Press 4 for division")

choice = int(input("Enter the use4rs choice :"))

if choice == 1:
    result = num1 + num2
    print(result)
elif choice == 2:
    result = num1 - num2
    print(result)
elif choice == 3:
    result = num1 * num2
    print(result)
elif choice == 4:
    result = num1 / num2
    print(result)
else:
    print("Invalid choice")
```
